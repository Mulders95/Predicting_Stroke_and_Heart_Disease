```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
stroke_df = pd.read_csv("Stroke/stroke_dataset_for_modeling.csv")
stroke_df
```

```python
# Sets of indices where stroke and heart risk are present
A = set(stroke_df[stroke_df['stroke'] == 1].index)
B = set(stroke_df[stroke_df['heart_risk'] == 1].index)

# Create the Venn diagram
plt.figure(figsize=(4, 5))
venn2_unweighted([A, B], set_labels=('Stroke', 'Heart Risk'), set_colors=("orange", "blue"), alpha=0.6)
plt.title("Stroke and Heart Risk Venn Diagram", fontweight="bold", fontsize=12, y = 0.94)

# Subtitle below the diagram
plt.figtext(0.5, 0.2, "Heart Risk implies the risk of stroke given either heart disease or hypertension",
            ha="center", fontsize=10, fontstyle="italic")
plt.show()
```

```python
heartdisease_df = pd.read_csv("HeartDisease/heartdisease_data_for_visual.csv")
heartdisease_df
```

```python
# Sets of indices where stroke and heart risk are present
A = set(heartdisease_df[heartdisease_df['heart_disease'] == 1].index)
B = set(heartdisease_df[heartdisease_df['stroke_risk'] == 1].index)

# Create the Venn diagram
plt.figure(figsize=(4, 5))
venn2_unweighted([A, B], set_labels=('Heart Disease', 'Stroke Risk'), set_colors=("orange", "blue"), alpha=0.6)
plt.title("Heart Disease and Stroke Risk Venn Diagram", fontweight="bold", fontsize=12, y = 0.94)

# Subtitle below the diagram
plt.figtext(0.5, 0.2, "Stroke Risk implies the risk of heart disease given either stroke or hypertension",
            ha="center", fontsize=10, fontstyle="italic")

plt.show()
```
