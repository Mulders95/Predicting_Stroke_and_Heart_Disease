```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
stroke_df = pd.read_csv("Stroke/stroke_dataset_for_modeling.csv")
stroke_df
```

```python
# Sets of indices where stroke and heart risk are present
A = set(stroke_df[stroke_df['stroke'] == 1].index)
B = set(stroke_df[stroke_df['heart_risk'] == 1].index)

# Create the Venn diagram
plt.figure(figsize=(8, 8))
venn2_unweighted([A, B], set_labels=('Stroke', 'Heart Risk'), set_colors=("green", "blue"), alpha=0.6)
plt.title("Stroke and Heart Risk Venn Diagram", fontweight="bold", fontsize=14, y = 0.94)
plt.show()
```

```python
heartdisease_df = pd.read_csv("HeartDisease/heartdisease_data_for_visual.csv")
heartdisease_df
```
